For her analysis she wants to do a survival analysis of the houses they retain in their portafolio and their overall qualifications

```python
def survival(data, group_field, time_field, event_field):
    kmf = KaplanMeierFitter()
    results = []

    for i in data[group_field].unique():
        group = data[data[group_field]==i]
        T = group[time_field]
        E = group[event_field]
        kmf.fit(T, E, label=str(i))
        results.append(kmf.survival_function_)

    survival = pd.concat(results, axis=1)
    front_fill = survival.fillna(method='ffill')
    back_fill = survival.fillna(method='bfill')
    smoothed = (front_fill + back_fill) / 2
    return smoothed
```

```python
rates= survival(Raffles_houses,'OverallQual','Age_house','H_act_inac')

rates.iplot(kind='line', xTitle='Age', yTitle='tasa de retencion',
            title='Retencion por genero')
```
