She does a survival analysis of the time the houses stay in the portafolio since they are sold based on the vintage category of the house.

```python
def survival(data, group_field, time_field, event_field):
    kmf = KaplanMeierFitter()
    results = []

    for i in data[group_field].unique():
        group = data[data[group_field]==i]
        T = group[time_field]
        E = group[event_field]
        kmf.fit(T, E, label=str(i))
        results.append(kmf.survival_function_)

    survival = pd.concat(results, axis=1)
    front_fill = survival.fillna(method='ffill')
    back_fill = survival.fillna(method='bfill')
    smoothed = (front_fill + back_fill) / 2
    return smoothed
```

```python
rates= survival(enterprise_houses,'Vintage','Vintage in Portafolio','Stat')

rates.iplot(kind='line', xTitle='Vintage in portafolio', yTitle='retention rate',
            title='Portafolio retention')
```

```python
# With this analysis she saws that the houses that stay more time at the portafolio are the old houses. For example at the fourth year there is 43% of new houses and 49.26% of the old ones in the portafolio.
```

Supose that in the past month you bought a raffle ticket. And today the raffles director calls you and informs that you win a "Big House" and in the next days he will call you to provide more details about the house. 

Based on the terms and conditions of your ticket it mentions that a big house is the one that LotArea is greater or equal to 12,750m.And also mentions that the raffle houses are based in College Creek and Old Town and that you can consult the list on internet (The internet list mentions that all the houses mentioned there are available for the raffle). Now you want to know the probability that your house is located in College Creek(Loc_1) or Old Town(Loc_8), based that it is a "big house".

```python
Loc_1["Size"]=np.where(Loc_1['LotArea']>=12750, 'Big', 'Small')
Loc_8["Size"]=np.where(Loc_8['LotArea']>=12750, 'Big', 'Small')
```

```python
Loc_1.groupby(["Size"]).agg({"Id":["count"]}).reset_index()
```

```python
Loc_8.groupby(["Size"]).agg({"Id":["count"]}).reset_index()
```

```python
neighborhoods=[1/2,1/2]
prob_big_house=[2/25,5/113]

def bayes(neighborhoods,prob_big_house):
    marginal = sum(np.multiply(neighborhoods,prob_big_house))
    num = np.multiply(neighborhoods,prob_big_house)
    posteriori = np.divide(num, marginal)
    return posteriori
```

```python
bayes(neighborhoods,prob_big_house)
```

```python
# In conclusion after applying the Bayes Theorem determines that the probability that is a big house and its from CollgCr is 0.64387464 and the probability that is a big house and its from Old Town is 0.35612536 so now is very excited to know what result will be. 
```
